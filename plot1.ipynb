```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Re = 1000
Mx = 32
n = 100

u = pd.read_csv(f"Re_{Re}_{Mx}/u_{n}.csv", header=None).to_numpy()[1:-1, :]
v = pd.read_csv(f"Re_{Re}_{Mx}/v_{n}.csv", header=None).to_numpy()[:, 1:-1]

dx = dy = 1/32

div = np.diff(u, axis=1)/dx + np.diff(v, axis=0)/dy
print("div: ", np.linalg.norm(div)/32)
plt.contourf(div, levels=3)
plt.colorbar()
```

```python
file_name = "RE_1000_128"
n = 30000
# CFD 결과 데이터 불러오기 (CSV 또는 Numpy 배열)
U = np.loadtxt(f"{file_name}/u_{n}.csv", delimiter=",")[1:-1, :]
V = np.loadtxt(f"{file_name}/v_{n}.csv", delimiter=",")[:, 1:-1]

U = (( U[:, 1:] + U[:, :-1] )/2)
V = (( V[1:, :] + V[:-1, :] )/2)
speed = np.sqrt(U**2 + V**2)

# CFD 격자 좌표 생성
Nx, Ny = U.shape
dx = 1/Nx
grid = np.linspace(dx, 1-dx, Nx)
X, Y = np.meshgrid(grid, grid)

plt.figure(figsize=(5, 5))
# Streamlines 플롯
plt.streamplot(X, Y, U, V, density=3, 
               linewidth= 1, #3*speed / speed.max(), 
               arrowsize=1, 
               cmap="jet")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0, 1)
plt.ylim(0, 1)

# plt.quiver(X, Y, U, V, 
#            angles='xy', 
#            scale_units='xy', 
#            scale=1,        # 크기 스케일 조절
#            cmap='jet')

```
