```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Re = 1000
Mx = 4
n = 1
name = "phi"

u_1 = np.array(pd.read_csv(f"Re_{Re}_{Mx}/{name}_{n}.csv", header=None))
u_2 = np.array(pd.read_csv(f"{name}_{n}.csv", header=None))
error = u_1 - u_2

plt.figure(figsize=(12, 3))  # 전체 그림 크기 설정

plt.subplot(1, 3, 1)
plt.contourf(u_1)
plt.colorbar()

plt.subplot(1, 3, 2)
plt.contourf(u_2)
plt.colorbar()

plt.subplot(1, 3, 3)
plt.contourf(error)
plt.colorbar()

plt.tight_layout()

plt.show()
```

```python
Re = 1000
Mx = 4
n = 1

u = np.array(pd.read_csv(f"Re_{Re}_{Mx}/u_{n}.csv", header=None))[1:-1, :]
v = np.array(pd.read_csv(f"Re_{Re}_{Mx}/v_{n}.csv", header=None))[:, 1:-1]

dx = dy = 1/4

div = np.diff(u, axis=1)/dx + np.diff(v, axis=0)/dy
print("div: ", np.linalg.norm(div)/4)
plt.contourf(div, levels=5)
plt.colorbar()
```
